```python
## DOCUMENTATION FOR TKINTER

# tkinter.filedialog
# (Common dialogs to allow the user to specify a file to open or save.)

# tkinter.messagebox
# (Access to standard Tk dialog boxes.)

# The tkinter.ttk
# (module provides access to the Tk themed widget set behavior from the code implementing its appearance.)
# WIDGETS(Button, Checkbutton, Entry, Frame, Label, LabelFrame, Menubutton, PanedWindow, Radiobutton, Scale and Scrollbar)


# PACKER OPTIONS
# The pack() method can be called with keyword-option/value pairs that control where the widget is to appear within its container, and how it is to behave when the main application window is resized.
# Here are some examples
# 1)anchor
# Anchor type. Denotes where the packer is to place each slave in its parcel.
# 2)ipadx and ipady
# A distance - designating internal padding on each side of the slave widget.
# 3)padx and pady ON DIFFRENT AXIS (X & Y)
# A distance - designating external padding on each side of the slave widget.
# 4)side
# Legal values are: 'left', 'right', 'top', 'bottom'.

# WIDGETS
# Widget is an element of Graphical User Interface (GUI) that displays/illustrates information
# NOTE IT STARTS WITH A CAPITAL LETTER
# Widgets	Description
# Label	It is used to display text or image on the screen
# Button	It is used to add buttons to your application
# Canvas	It is used to draw pictures and others layouts like texts, graphics etc.
# ComboBox	It contains a down arrow to select from list of available options
# CheckButton	It displays a number of options to the user as toggle buttons from which user can select any number of options.
# Radio Button	It is used to implement one-of-many selection as it allows only one option to be selected
# Entry	It is used to input single line text entry from user
# Frame	It is used as container to hold and organize the widgets

# LABEL WIDGET
# Label() widget use to display text that users can’t able to modify.
# root is the name of the window
# text which we display the title of the label
# font in which our text is written
# pack() organized widget in block

# BUTTON WIDGET
# Button() widget used to display button on the window.
# text- which we display on the label
# font- in which the text is written
# bg- sets the background color
# command- is used to call the function

# GEOMETRY MANAGEMENT
# 1)pack()	The Pack geometry manager packs widgets in rows or columns.
# 2)grid()	The Grid geometry manager puts the widgets in a 2-dimensional table. 
# The master widget is split into a number of rows and columns, and each “cell” in the resulting table can hold a widget.
# 3)place()	The Place geometry manager is the simplest of the three general geometry managers provided in Tkinter. 
# It allows you explicitly set the position and size of a window, either in absolute terms, or relative to another window.

# root.mainloop() is a method that executes when we want to run the program.
```

```python
# INSTALLATION OF IMPORTANT PACKAGES
import time
import requests
from tkinter import *
from pytube import YouTube
from tkinter import ttk,filedialog,messagebox
```

```python
# CREATING A DISPLAY WINDOW
# *Tk() used to initialize tkinter to create display window
# *geometry() used to set the window’s width and height
# *resizable(0,0) set the fix size of window
# *title() used to give the title of window

def main_page():
    root=Tk()
    root.title("Mugambi's Youtube Downloader")
    root.geometry("500x320")
    root.resizable(width=False,height=False)
    global frame1 ##allows us to modify the variable outside of the current scope.
    
    #The frame is anchored to the North and has a padding of 10 units on the x-axis.
    top = Frame(root)
    top.pack(side=TOP,anchor=N,padx=10)
    frame1=Frame(top)
    frame1.pack(side=TOP,anchor=N,padx=10)
    
    
    ## Check internet connection
    def check_internet():
        url="https://www.google.com"
        timeout =5
        try:
            request =requests.get(url, timeout=timeout)
            return True
        except:
            return False
        
    if check_internet() == True:
        lbl =Label(frame1, text="Internet Connected", font="San-serif 14 bold", foreground="green").pack()
    elif check_internet() == False:
        lbl =Label(frame1, text="Internet Disconnected", font="San-serif 14 bold", foreground="red").pack()
    
    
    lbl =Label(frame1, text="Download Youtube Videos for free", font="San-serif 14 bold").pack()

    
    frame2=Frame(top)
    #adds a 10-pixel padding along the y-axis
    frame2.pack(pady=10)
    
    #Creating the tkinter Variables where the video is stored from the entry
    #It is used to represent the value of an entry widget, for example, in a GUI interface.
    video_link= StringVar()
    
    lbl =Label(frame2, text="Paste Your Link Here", font="San-serif 14 bold").pack()
    entry= Entry(frame2,width=60, textvariable=video_link).pack()
    
    
    frame3=Frame(top)
    frame3.pack(pady=10)
    
    download_path= StringVar()
    
    def select():
        path=filedialog.askdirectory(initialdir="YOUR DIRECTORY PATH", title="Save Video")
        download_path.set(path)
    
    def download_video():
        url=video_link.get()
        folder=download_path.get()
        get_video= YouTube(url)
        get_stream= get_video.streams.first()
        get_stream.download(folder)
        
        messagebox.showinfo("Successful", "Downloaded and saved in\n"+folder)
    
    lbl =Label(frame3, text="Choose path to Save:", font="San-serif 14 bold").grid(row=0,column=0)
    selectbtn =ttk.Button(frame3, text="select", command=select).grid(row=0,column=1)
    lbl =Label(frame3, text="Path Selected:", font="San-serif 14 bold").grid(row=1,column=0)
    lbl =Label(frame3, text="", font="San-serif 14 bold", textvariable =download_path).grid(row=1,column=1)
    
    frame4=Frame(top)
    frame4.pack(pady=10)
    
    downloadbtn= Button(frame4, text="Download", command=download_video ,font="San-serif 14 bold", foreground="white", background="red").grid(row=0,column=0)
    
    
    check_internet()
    root.mainloop()
```

```python
main_page()
```
